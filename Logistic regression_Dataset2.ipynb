```python
df["Time"] = pd.to_datetime(df["Transaction Date and Time"], format='%Y-%m-%d %H:%M:%S')
```

```python
times = pd.Series(['00:00:00','06:00:00','12:00:00','18:00:00'])
```

```python
times = pd.to_datetime(times)
times = times.dt.time
times
```

```python
recorded_times = df.Time.dt.time
recorded_times
```

```python
def categorize_time(x):
    if times[0] <= x < times[1]:
        return 1
    if times[1] <= x < times[2]:
        return 2
    if times[2] <= x < times[3]:
        return 3
    else:
        return 4
```

```python
df["Recorded_times"] = recorded_times
```

```python
df.drop(columns=["Time"])
```

```python
categorized_times = df["Recorded_times"].apply(categorize_time)
df["Categorized_times"] = categorized_times
df
```

```python
def categorize_transaction_purpose(x):
    if 1 <= x <= 1499 or 4800 <= x <= 4999 or 3300 <= x <= 3499 or 3500 <= x <= 3999 or 5600 <= x <= 5699:
        return 1
    if 4000 <= x <= 4799 or 5000 <= x <= 5599 or 7300 <= x <= 7999 or 8000 <= x <= 8999 or 9000 <= x <= 9999 or 3000 <= x <= 3299:
        return 2
#         if i in "health fitness care":
#             return 3
    return 4
```

```python
categorized_transaction_purpose = df["Merchant Category Code (MCC)"].apply(categorize_transaction_purpose)
df["Categorized_trasaction_purpose"] = categorized_transaction_purpose
df
```

```python
# Load data
X = df[["Transaction Amount","Categorized_times", "Categorized_trasaction_purpose"]]

# Extracting Target / Class Labels
y = df['Fraud Flag or Label']
```

```python
from sklearn.model_selection import train_test_split
# Train Test split 
X_train, X_test, y_train, y_test = train_test_split(X,y, 
                                     train_size = 0.7, 
                                     test_size = 0.3, 
                                     random_state = 42) 
```

```python
from sklearn.linear_model import LogisticRegression
# Create an instance of Logistic Regression Classifier and fit the data.
clf = LogisticRegression(C=1e5)
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
#evaluation on training data

y_pred = clf.predict(X_train)

print(confusion_matrix(y_train, y_pred))
print(classification_report(y_train, y_pred))
print(accuracy_score(y_train, y_pred))
```

```python
#evaluation on test data

y_pred = clf.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
