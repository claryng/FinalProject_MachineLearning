```python
def categorize_time(x):
    if times[0] <= x < times[1]:
        return 1
    if times[1] <= x < times[2]:
        return 2
    if times[2] <= x < times[3]:
        return 3
    else:
        return 4
```

```python
df["Recorded_times"] = recorded_times
```

```python
categorized_times = df["Recorded_times"].apply(categorize_time)
```

```python
df["Categorized_times"] = categorized_times
```

```python
def categorize_transaction_purpose(x):
    x = x.split("_")
    for i in x:
        if i in "kid kids pet pets shopping shop daily clothes household home grocery":
            return 1
        if i in "education entertainment transportation dining travel":
            return 2
        if i in "health fitness care":
            return 3
    return 4
```

```python
categorized_transaction_purpose = df["category"].apply(categorize_transaction_purpose)
```

```python
df["Categorized_trasaction_purpose"] = categorized_transaction_purpose
```

```python
# Get DoB
df["Birth Year"] = pd.to_datetime(df["dob"], format='%Y-%m-%d').dt.year
```

```python
df["Birth Year"]
```

```python
def categorize_dob_into_generation(x):
    if 1901 <= x <= 1927:
        return 0 # Greatest 
    elif 1928 <= x <= 1945:
        return 1 # Silent 
    elif 1946 <= x <= 1964:
        return 2 # Boomers
    elif 1965 <= x <= 1980:
        return 3 # Gen X
    elif 1981 <= x <= 1996:
        return 4 # Gen Y
    else:
        return 5 # Gen Z
```

```python
categorized_dob = df["Birth Year"].apply(categorize_dob_into_generation)
min(df["Birth Year"])
```

```python
df["Categorized DOB"] = categorized_dob
```

```python
df
```

```python
def categorize_gender(x):
    if x == "M":
        return 0 # Male
    else:
        return 1 # Female
```

```python
categorized_gender = df["gender"].apply(categorize_gender)
```

```python
df["Categorized Gender"] = categorized_gender
```
